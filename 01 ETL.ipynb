```python
import requests
import pandas as pd
import re

def import_info(id):

    #Funcion para descargar los excel de la fuente
    # Tu clave de API
    api_key = "WHZwSdXyIwvk138GPBmgvpFp5tCtr2t51d0rLPho"

    # URL base de la API de Junar
    base_url = "http://api.datosabiertos.enacom.gob.ar/api/v2"

    # URL para obtener el conjunto de datos específico
    dataset_url = f"{base_url}/datasets/{id}.json?auth_key={api_key}"

    try:
        # Realizar una solicitud GET al conjunto de datos específico
        response = requests.get(dataset_url)

        # Verificar si la solicitud fue exitosa (código de estado 200)
        if response.status_code == 200:
            # Parsear la respuesta JSON
            dataset_data = response.json()
            # Obtener el enlace de descarga (download_url) del conjunto de datos
            download_url = dataset_data.get("download_url")
            title = dataset_data.get("title")
            title = re.sub(r'[\\/:*?"<>|]', ' ', title)
            with requests.get(download_url, stream=True) as response:
                response.raise_for_status()
                with open(f"dataset/{title}.xlsx", 'wb') as file:
                    for chunk in response.iter_content(chunk_size=8192):
                        file.write(chunk)
            # Cargar el archivo Excel en un DataFrame de pandas
            excel_file = pd.ExcelFile(f"dataset/{title}.xlsx")
            nombres_de_hojas = excel_file.sheet_names

            return print(f"El archivo excel \"{title}.xlsx\" ha sido creado con exito  \
                    \nEste archivo tiene la siguiente lista de hojas:  \
                    \n{nombres_de_hojas}")

        else:
            print(f"Error al hacer la solicitud: Código de estado {response.status_code}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"Error al hacer la solicitud: {e}")
        return None

# Ejemplo de uso:
# df = importar_info("CONEC-AL-SERVI-DE-INTER")
```

## Transformación

Una vez tenemos los archivos que necesitamos empezamos con la transformación de los archivos al formato csv, para ello crearemos un función que buscara archivos excel en la carpeta para convertir sus hojas en archivos csv.

```python
import pandas as pd
import os

def excel_a_csv(directorio_excel, directorio_csv):

    #La función convierte las hojas de cálculo de archivos Excel en archivos CSV con nombres específicos.

    # Obtenemos la lista de archivos Excel en el directorio de entrada y nos aseguramos que sea tipo excel
    archivos_excel = [archivo for archivo in os.listdir(directorio_excel) if archivo.endswith('.xlsx')]

    # Procesaremos cada archivo Excel
    for archivo_excel in archivos_excel:
        # Se lee archivo Excel en un DataFrame de pandas
        # El argumento sheet_name=None permite cargar todas las hojas de cálculo en un diccionario.
        df_excel = pd.read_excel(os.path.join(directorio_excel, archivo_excel), sheet_name=None)

        # Iteraremos a través de las hojas de cálculo en el archivo Excel
        for hoja, df in df_excel.items():
            # Creamos el nombre del archivo csv combinando el nombre del archivo Excel y el nombre de la hoja
            nombre_csv = f"{os.path.splitext(archivo_excel)[0]} - {hoja}.csv"
            
            # Ponemos la ruta completa al archivo csv de salida
            ruta_csv = os.path.join(directorio_csv, nombre_csv)
            
            # Guardamos la hoja de cálculo actual como archivo csv
            df.to_csv(ruta_csv, index=False)

            # Imprimimos un mensaje de progreso
            print(f"Archivo CSV guardado: {nombre_csv}")
```
